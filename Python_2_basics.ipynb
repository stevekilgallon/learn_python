## Indentation "The rule of 4" 

Python has some strict rules on how you need to set out your code.

In Excel an example of a loop is given below

For i = 1 To 6
    Cells(i, 1).Value = 100
Next i

The loop starts with the For i .... and ends with Next i

In Python the for loop is diffent, see below. 
It starts with for i in range(0,10): and python uses "spaces" to deternine what is inside the loop.

````python
for i in range(0,10):
    print (i,"inside loop") #inside loop
print("outside loop")
````

Run the two cells below to see the effects of indentation in python.

````python
days_of_the_week = ["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]

for day in days_of_the_week:
    statement = "Today is " + day
    
    print (statement)
````

````python
days_of_the_week = ["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]

for day in days_of_the_week:
    statement = "Today is " + day
    
print (statement)
````

Both Loops start with a colon ":"       **"for day in days_of_the_week:"**
In the First cell the  "print (statement)" is inside the loop, but in the second cell it is outside the loop.
Python uses indentations (spaces) to idetify where blocks of code  begin and end.

**If you get an indentation error this is why!**

```python
if a==1:
    print(a)
    if b==2:
        print(b)
print('end')
```

In the above code, the first and last line of the statement is related to the same section because there is no indentation in front of them. So after executing first "if statement", the Python interpreter will go into the next statement. If the condition is not true, it will execute the last line of the statement.

**By default, Python uses four spaces for indentation**

**More Examples** <br>

**Example 1**<br>
```python

def my_function():     # This is a function definition. Note the colon (:)<br>
    a = 2              # This line belongs to the function because it's indented<br>
    return a           # This line also belongs to the same function<br>
print(my_function())   # This line is OUTSIDE the function block<br>

**Example 2**

if a > b:             # If block starts here<br>
    print(a)          # This is part of the if block<br>
else:                 # else must be at the same level as if<br>
    print(b)          # This line is part of the else block<br>
```
